```python
import pandas as pd

# Load the dataset
df = pd.read_csv("london_rental_affordability.csv")

# Preview the data
print(df.head())

# Sort by affordability
df_sorted = df.sort_values("Affordability_Ratio")

# Optional: Visualize as bar chart
import plotly.express as px
fig = px.bar(df_sorted, x='Borough', y='Affordability_Ratio',
             title='Affordability Ratio by Borough (Rent / Income)',
             labels={'Affordability_Ratio': 'Rent-to-Income Ratio'})
fig.show()
```

```python
import plotly.express as px

# Sort by affordability (lower is better)
df_sorted = df.sort_values("Affordability_Ratio")

# Create bar chart
fig = px.bar(
    df_sorted,
    x="Borough",
    y="Affordability_Ratio",
    title="London Boroughs by Rent-to-Income Ratio",
    labels={"Affordability_Ratio": "Rent / Income Ratio"},
    text_auto=".2f"
)

fig.show()
```

```python
import json
import plotly.express as px
import pandas as pd

# Load your data
df = pd.read_csv("london_rental_affordability.csv")

# Load GeoJSON
with open("london_boroughs.json") as f:
    geojson = json.load(f)

# Plot the map
fig = px.choropleth_mapbox(
    df,
    geojson=geojson,
    locations="Borough",
    featureidkey="properties.name",
    color="Affordability_Ratio",
    color_continuous_scale="YlOrRd",
    mapbox_style="carto-positron",
    zoom=9,
    center={"lat": 51.5074, "lon": -0.1278},
    opacity=0.7,
    labels={"Affordability_Ratio": "Rent / Income"}
)

fig.update_layout(
    title_text="London Boroughs: Rent-to-Income Affordability",
    margin={"r":0,"t":40,"l":0,"b":0}
)

fig.show()
```
